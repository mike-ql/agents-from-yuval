## Code-Usage of a local LLM
We'll use **curl** and python libraries to connect to our local LLM model.  
On Windows, make sure you type **curl.exe**, otherwise curl will be an alias for **Invoke-WebRequest**, which has a completely different syntax than the standard Linux/Unix curl.

### Call model using curl (Windows: curl.exe)
- Type or copy the following command:  
```curl.exe http://localhost:11434/api/generate -d '{\"model\": \"gemma3:12b\", \"prompt\": \"Why is the sky blue?\", \"stream\": false}'```

### Call a local model using openai library
- I have already downloaded and installed the ollama tool
- I have downloaded the gemma3:12b model (too large for my GPU, but ok inside my laptop RAM)

````python
import openai

client = openai.OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama" # Required but ignored
)

response = client.chat.completions.create(
    model="gemma3:12b",
    messages=[{"role": "user", "content": "Hello Llama!"}]
)

print(response.choices[0].message.content)
````
