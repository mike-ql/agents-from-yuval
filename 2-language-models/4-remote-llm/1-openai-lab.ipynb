- connect to [openai platform](https://openai.com/api/)

- Click on **getting started**
- Choose OS and click on **Create ab API key**
- A good place to store the file (Windows) would be in the $PROFILE init file
- Click on **Create new secret key**
- Make sure you copy is - then sabe it like this:  
setx OPENAI_API_KEY "your_api_key_here"
- Run (reload) it like this:  
```. $PROFILE```
