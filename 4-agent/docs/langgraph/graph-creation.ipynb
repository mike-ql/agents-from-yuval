#### Generic Graph Creation

- [StateGraph](https://reference.langchain.com/python/langgraph/graphs/#langgraph.graph.state.StateGraph)  
This is a class, so when we instantiate it we get a "graph definition" object. We'll have to "compile" this one to get an executable graph.

```python
from langgraph.graph import StateGraph

from typing import TypedDict

class SimpleState(TypedDict):
    # This is the only "variable" our graph tracks
    text: str

graph = StateGraph(SimpleState)
print(type(graph))

```
