#### Pay Bill Agent - Exercise
- Write an agent that helps you pay a bill.
- The agent's logic is:
  - Look at the message history.
  - Check what items are still missing: (account number, sum, month, email)
  - If something is missing, ask for it.
  - If everything is present, "book" the trip.
- Construct the output as a json (use a schema)
- Make sure you accumulate messages, and not rewriting the last message each time.
