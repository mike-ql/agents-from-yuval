### Exercise: The Office Welcome-Back Mission

##### **The Scenario**
Suzy from the QA team is returning from maternity leave, and the office wants to mark the occasion. You are designing a **Gift Coordinator Agent** tasked with a simple but high-stakes mission: gather enough funds from the team to get her a great gift. If the collection falls short, the agent must pivot to a thoughtful backup plan so she doesn't return to an empty desk.

##### **The Objective**
Your task is to build the decision-making logic for an agent that manages a limited "collection" window. The agent must decide whether to keep asking for donations or to wrap up the campaign based on the total amount gathered, (maximum two collection attempts).

##### **The Workflow Logic**
The agent must navigate the following sequence to determine Suzy’s welcome-back surprise:

1. **The First Collection:** The agent starts by reaching out to the department. This action returns a random contribution amount.
2. **The Threshold Check:**
  - **Success:** If the amount collected is **$60 or more**, the agent immediately stops and proceeds to the "Success Path."
  - **Insufficient Funds:** If the amount is **less than $60**, the agent must decide if it has the "energy" for one more round.
3. **The Final Push (Optional):** If the first round was short, the agent can trigger a second and final collection attempt. The new funds are added to the previous total.
4. **Final Decision Node:** 
  - **Path A (The Gift):** If the total reaches or exceeds the **$60 goal** at any point, the agent triggers the "Purchase Gift" action.
  - **Path B (The Card):** If, after the second attempt, the total is still **under $60**, the agent must stop and trigger the "Send Congratulations Card" action instead.
5. **Completion:** The agent reports the final outcome—either the gift or the card—and closes the task.

**Key Requirements**

* **Loop Constraints:** The agent is strictly forbidden from calling the collection action more than twice. It must track its "attempt count" accurately.
* **State Accumulation:** The agent must be able to add new data (the second collection) to its existing knowledge (the first collection) without losing the original value.
* **Conditional Branching:** The logic must clearly distinguish between "Having enough money" and "Running out of turns."
