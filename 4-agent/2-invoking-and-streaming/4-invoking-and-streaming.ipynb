## Invoking and Streaming


```python
from dotenv import load_dotenv   # from the python-dotenv library - load the OpenAI API key from file
from langchain_openai import ChatOpenAI

load_dotenv()     # This will load the OpenAI API key from the .env file
```

```python
llm = ChatOpenAI( 
    model="gpt-4o", # or "gpt-3.5-turbo"
    temperature=0,
)
print(f"{llm.__class__.__module__}.{llm.__class__.__name__}")
print(llm._llm_type)
```

```python
# This will take some time because we have to wait for the complete response to be ready
response = llm.invoke('write a 200 words about flowers')
print(response)
```

```python
# Now, stream will respond with an iterator
resp_iter  = llm.stream('write a 200 words about butterflies')
for chunk in resp_iter:
    print(chunk.content, end='|\n')
```
