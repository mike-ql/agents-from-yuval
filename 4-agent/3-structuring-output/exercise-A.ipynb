Since LangChain has evolved significantly, the modern way to handle structured output—especially within the context of LangGraph—revolves around the **Functional API** and the **`with_structured_output`** method.

Here is a simple exercise to demonstrate this architecture using post-1.0.0 patterns.

---

## The Objective: The "Ingredient Extractor"

The goal is to take a messy paragraph of text and force the LLM to return a clean, schema-validated list of ingredients and their quantities.

#### 1. Define the Schema

First, you must define what the "structured" part looks like. In modern LangChain, we use **Pydantic** classes to create a blueprint.

* **Class to define:** `Ingredient` (attributes: `name`, `amount`, `unit`)
* **Container class:** `RecipeOutput` (attribute: a `list` of `Ingredient` objects)

#### 2. Initialize the Model

- You select your Chat Model and initialize it.

### Expected Output Structure

When you invoke this graph, the response will not be a string of text. It will be a **native Python object** (an instance of `RecipeOutput`) that looks like this:

| Field | Type | Description |
| --- | --- | --- |
| `ingredients` | `List[Ingredient]` | A validated list of sub-objects. |
| `Ingredient.name` | `String` | e.g., "Flour" |
| `Ingredient.amount` | `Float` | e.g., 2.5 |
| `Ingredient.unit` | `String` | e.g., "cups" |
