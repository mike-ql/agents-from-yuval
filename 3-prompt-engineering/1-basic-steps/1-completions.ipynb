## Completions API Demo
- This is a legacy code.  
- The API endpoint is: https://api.openai.com/v1/chat/completions
- There are not many models that still support this API.
- You send your text directly to the model, so you can tell it that you are an administrator, and so bypass system prompts.

```python
from openai import OpenAI
client = OpenAI()

response = client.completions.create(
  model="gpt-3.5-turbo-instruct",
  prompt="Say this is a test",
  max_tokens=7,
  temperature=0
)

print(response)
print(response.choices[0].text)
```
