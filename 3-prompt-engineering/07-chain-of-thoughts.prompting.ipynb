## Chain of Thoughts pattern
In the world of AI prompting, Chain of Thought (CoT) is essentially asking the model to "show its work."

Instead of jumping straight from a question to an answer—which can lead to "hallucinations" or logic errors—CoT encourages the AI to walk through a series of intermediate reasoning steps. It’s the difference between a student shouting out "42!" and a student writing out the entire equation on the chalkboard.

link:  
[Chain-of-Thought Prompting Elicits Reasoning in Large Language Models](https://arxiv.org/abs/2201.11903)

#### How It Works
Standard prompting provides a "Input -> Output" structure. CoT adds a "Reasoning" layer in the middle.

1. Zero-Shot CoT
This is the simplest version. You don't even have to give examples; you just add a specific phrase to your prompt.
The Magic Phrase:  
**Let’s think step by step.**

Why it works: This triggers the model to break the problem into smaller, logical chunks before committing to a final answer.

2. Few-Shot CoT
You provide the AI with a few examples that include the reasoning process.


#### Example 1
**Prompt:**
```
My car speed is 100 km/hour.
I am driving for 1.5 hours, then take a 30 minutes break.
Then I drive for 2 more hours, and I reach my destination.
How far was my destintion?
Let's think step by step.
```
**Response:**  
To find the total distance to your destination, we can break the journey down into stages:  
First leg of the drive:  
You drove at a speed of $100 \text{ km/h}$ for $1.5 \text{ hours}$. $$\text{Distance} = \text{Speed} \times \text{Time}$$  
$$\text{Distance}_1 = 100 \text{ km/h} \times 1.5 \text{ h} = 150 \text{ km}$$  
The break:  
You took a $30\text{-minute}$ break.  
Since the car was not moving during this time, the distance covered is $0 \text{ km}$.  
Second leg of the drive:  
You drove for another $2 \text{ hours}$ at the same speed ($100 \text{ km/h}$).  
$$\text{Distance}_2 = 100 \text{ km/h} \times 2 \text{ h} = 200 \text{ km}$$  
Total distance:  
Summing the distances from both legs:$$\text{Total Distance} = 150 \text{ km} + 200 \text{ km} = 350 \text{ km}$$Your destination was $350 \text{ km}$ away.

#### Example 2 (few-shot version)
**Prompt:**
```
Question: I have 3 apples. I buy 2 more crates with 10 apples each. How many do I have?
Reasoning: I start with 3 apples. 2 crates of 10 is 20 apples. 3 + 20 = 23.
Answer: 23.

I have 33 Apples. I give2 apples to each of my 15 friends.
How many Apples do I have left?
```
**Response:**
Reasoning: I start with 33 apples. I give 2 apples to each of my 15 friends, which is  apples given away. .

Answer: 3.
