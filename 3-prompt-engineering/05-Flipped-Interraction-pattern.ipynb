## The Flipped Interaction pasttern
The Flipped Interaction pattern is a prompt engineering technique that reverses the typical "user-asks-AI-answers" dynamic. Instead of you providing all the context and hoping the AI hits the mark, you instruct the AI to ask you questions until it has enough information to achieve a specific goal.

### How it Works
In a standard interaction, you might provide a long, complex prompt. If you miss a detail, the AI "hallucinates" or makes assumptions to fill the gaps.
With the Flipped Interaction, the AI pauses its generation phase and enters an "information gathering" phase.  
It will continue to probe for details until a predefined threshold is met.

### The Basic Structure
To trigger this pattern, your prompt should include these three components:
- The Goal: What you want to achieve (e.g., "Write a meal plan").
- The Instruction: Explicitly tell the AI to ask questions.
- The Stopping Condition: Tell it when to stop asking and start acting.

### Why Use This Pattern?
- Precision - Prevents the AI from making incorrect assumptions about your needs.
- Efficiency - You don't have to write a massive "mega-prompt" upfront; you just respond to prompts.
- Customization - The output is tailored specifically to the data points the AI knows it needs.
- Reduced Cognitive Load  - It’s often easier to answer five specific questions than to write a one-page brief from scratch.

#### Example 1
**prompt:**
```
I want you to help me buy a new house and move in.
Please ask me several questions to get the information you need, and I will answer them.
When you feel you have the answers, go ahead and plan my task.
```
**response:**
Buying a home and orchestrating a move is a massive undertaking, but we can break it down into a streamlined project. To build a plan that actually works for your life, I need to get a sense of your "must-haves" and your timeline.

Please answer these questions to get us started:

### **1. The Search (The "Where" and "What")**

* **Location:** Do you have specific neighborhoods or cities in mind, or are you looking for a certain vibe (e.g., urban, quiet suburbs, proximity to work)?
* **The House:** How many bedrooms/bathrooms do you need? Are there deal-breakers (e.g., must have a backyard, home office, or no stairs)?

### **2. The Financials**

* **Budget:** Do you have a firm price range or a maximum monthly payment you're aiming for?
* **Status:** Have you already been pre-approved for a mortgage, or do we need to start with the "getting your ducks in a row" phase?

### **3. The Logistics (The Move)**

* **Current Situation:** Are you selling a home at the same time, or are you currently renting?
* **Timeline:** What is your "dream" move-in date? Are you in a rush, or do you have several months to play with?

### **4. The Heavy Lifting**

* **DIY vs. Pro:** Do you plan on packing and moving everything yourself, or are you looking to hire full-service movers?

---

**Once you provide these details, I’ll build a customized roadmap for you, ranging from the initial house hunt to the day you unpack your first box.**

How would you like to tackle the first set of questions?


**Prompt:**
```
Location: Kochav Ya'ir
The house: 5 bedrooms, 2 bathrooms, yard area of at least 500square meters
Budget:  4-7 million IS
Status: no pre-approved mortage
Current Situation: Selling my current home
Timeline:   ~ 1 year from now
```
**Response:**
Kochav Ya'ir is a beautiful choice—it's known for its high quality of life and "village" feel, but it's also a tight market. With a 1-year timeline and the need to sell your current home simultaneously, we need to be very tactical. Your budget of **4–7 million NIS** is realistic for the area, but 5-bedroom homes with large plots (500sqm+) often sit at the higher end of that range.

Here is your 12-month strategic plan to transition to your new home.

---

## **Phase 1: Financial & Legal Setup (Months 1–2)**

Since you are selling and buying at once, your cash flow depends on the timing of these two deals.

* **Get an "Ishur Ekroni" (Mortgage Pre-approval):** Even if you don’t have a house yet, you need this to know exactly what the bank will lend you. In Israel, this usually lasts 24–90 days but can be easily renewed.
* **Appraise Your Current Home:** Hire an appraiser (*shamai*) now. You need to know exactly how much equity you’ll have for the new purchase.
* **Hire a Real Estate Attorney:** In Israel, the lawyer handles the "closing." You’ll want one who specializes in "back-to-back" transactions to ensure your selling contract aligns with your buying contract.

## **Phase 2: The Dual Market Entry (Months 3–6)**

* **List Your Current Home:** Most sellers in your position find a buyer first but include a "long closing" (e.g., 8–10 months) to give themselves time to find the Kochav Ya'ir house.
* **The Hunt in Kochav Ya'ir:** 500sqm lots are the standard in the older parts of Kochav Ya'ir. Start visiting now to see the difference between "original" houses and those that have been renovated.
* *Note:* At **7M NIS**, you can find very nice 5-bedroom homes. At **4-5M NIS**, you may find houses that require a significant renovation budget.



## **Phase 3: The Contract Bridge (Months 6–9)**

* **Synchronize Payments:** Your lawyer will structure your sale so that the first large payment you receive from your buyer can cover the down payment (usually 15–20%) for the house in Kochav Ya'ir.
* **Due Diligence:** Before signing in Kochav Ya'ir, have your lawyer check the *Tabu* (Land Registry) and ensure the 5th bedroom was built with a legal permit (common issue in older private homes).

## **Phase 4: Logistics & Moving (Months 10–12)**

* **Book the "Hovala" (Movers):** For a 5-bedroom house, you’ll need a crew of at least 4–5 people. Book 2 months in advance if your move falls near the summer or holidays.
* **School/Local Registration:** Since you’re moving to a specific community, contact the *Moatza* (local council) early to register kids for schools or to handle the "Toshav" (resident) stickers for your cars.
