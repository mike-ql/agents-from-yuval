## Few-Shot pattern
Few-shot prompting is a technique where you provide a generative model with a small set of examples (shots) to demonstrate the specific task, format, or tone you want it to emulate.

While large language models are often capable of "Zero-shot" performance (following a direct instruction without examples), few-shot prompting is the "show, don't just tell" approach that significantly improves accuracy for complex or niche tasks.

### How It Works
The pattern typically consists of three components:
- The Task Description: (optional) A brief instruction of what the model should do.
- The Examples: A series of input-output pairs.
- The New Input: The actual data you want the model to process.

### Why Use Few-Shot Prompting?
- Consistency: It forces the model to adhere to a specific output format (e.g., JSON, XML, or specific shorthand) without needing long, wordy instructions.
- Nuance: It helps the model pick up on subtle stylistic choices, like sarcasm, professional jargon, or a specific brand voice.
- Reduced "Hallucination": By providing a clear guardrail of examples, the model is less likely to deviate into creative but incorrect responses.
- Complex Logic: It allows you to demonstrate "Chain of Thought" reasoning by showing the model how to solve a problem step-by-step before asking it to solve a new one.
