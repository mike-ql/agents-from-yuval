## Combining Prompt patterns
Combining propt patterns is the practice of stacking different prompting techniques within a single interaction to get more accurate, structured, or creative results.

Think of prompt patterns like LEGO bricks. One brick (e.g., "Persona") gives you a character, but adding another (e.g., "Chain of Thought") gives that character a logical brain. By combining them, you reduce the "hallucination" rate and ensure the output matches your specific needs.

#### Example 1

Persona * CoT (Chain of Thought)

**Prompt:**
# ROLE
Act as a Michelin-star pastry chef with a deep understanding of food science and chemistry. 

# TASK
I need you to audit a failing cake recipe. 

# STEPS (Chain of Thought)
1. Analyze the ingredients and identify why the cake is currently collapsing in the center.
2. Explain the chemical relationship between the baking soda, the buttermilk, and the oven temperature.
3. Suggest specific technical adjustments to the mixing method.

# OUTPUT FORMAT
Provide the final, corrected recipe in the Markdown table format below.

---

# RECIPE TO AUDIT
- 2 cups All-Purpose Flour
- 1.5 cups Granulated Sugar
- 1 tbsp Baking Soda
- 1/2 tsp Salt
- 1 cup Buttermilk
- 1/2 cup Melted Butter
- 2 Large Eggs
- 1 tsp Vanilla Extract

**Instructions:** Mix everything in a bowl for 5 minutes until very smooth. Bake at 375°F for 30 minutes.


**Response:**

Bonjour. As a chef, I see exactly where your structure is failing. Baking is a precise dialogue between leavening gases and protein structures; currently, your recipe is shouting when it should be whispering, and your structure is collapsing under the weight of its own ambition.

### 1. Why the Center is Collapsing

The primary culprit is **leavener overdose**. You have 1 tablespoon (3 teaspoons) of baking soda for only 2 cups of flour. This creates a massive, rapid release of carbon dioxide. The cake rises aggressively before the egg and flour proteins have reached their "setting" temperature. Essentially, you are inflating a balloon that hasn't turned into rubber yet—it pops, and the center caves in.

### 2. The Chemistry: Soda, Acid, and Heat

In this recipe, we have a classic acid-base reaction:

* **The Reactants:** Baking soda (sodium bicarbonate) and Buttermilk (lactic acid).
* **The Reaction:** .
* **The Temperature Conflict:** At **375°F**, the exterior of the cake sets far too quickly. This traps the steam and  inside. Because the baking soda is so high, the internal pressure becomes too great for the weak protein network to hold, leading to a "volcano" effect followed by a total structural collapse.

### 3. Technical Mixing Adjustments

Your current "mix for 5 minutes" instruction is detrimental. Over-mixing develops excessive gluten, which makes a cake tough, but more importantly, it beats out the air you just tried to incorporate.

* **The "Muffin Method" approach:** Mix dry and wet separately.
* **The Fold:** Only combine until *just* moistened. Lumps are your friend; they disappear in the oven.
* **Timing:** Once buttermilk hits baking soda, the clock is ticking. You must get that batter into the oven within minutes before the bubbles dissipate.

### Revised Instructions

1. **Prep:** Preheat your oven to **350°F (175°C)**. The lower temperature allows the center to rise and set evenly with the edges.
2. **Dry Phase:** Whisk flour, sugar, baking soda, baking powder, and salt in a large bowl.
3. **Wet Phase:** In a separate jug, whisk buttermilk, melted butter, eggs, and vanilla.
4. **The Marriage:** Pour the wet into the dry. Fold gently with a spatula for **30–45 seconds**—just until no streaks of dry flour remain. **Do not use a mixer for 5 minutes.**
5. **Bake:** Pour into a prepared pan immediately. Bake for 30–35 minutes or until a skewer comes out clean.
