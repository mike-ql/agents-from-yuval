## hello ai example - download MNIST
The famous digit recognition example.
Here - we introduct torchvision and download the database.

```python
from torchvision import datasets
import torchvision.transforms as transforms
```

- **torchvision** is part of the [pytorch ecosystem](https://landscape.pytorch.org/).
  We need it to download the [MNIST Database](https://en.wikipedia.org/wiki/MNIST_database) that contains those 28x28 images of hand-written digits.
- The transform parameter is saying: "When you load each image, convert it to a PyTensor (a numerical format PyTorch understands)." 

```python
# Download MNIST dataset
print("Downloading MNIST dataset...")
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transforms.ToTensor())
```

```python
print(f"✓ Training set: {len(train_dataset)} images")
print(f"✓ Test set: {len(test_dataset)} images")
print(f"✓ Downloaded to: ./data/")
```

We can see some digits from the database:

```python
import matplotlib.pyplot as plt

# Get the first image from the dataset
first_image, label = train_dataset[1]

# Display it
plt.imshow(first_image.squeeze(), cmap='gray')
plt.title(f"Digit: {label}")
plt.show()
```
